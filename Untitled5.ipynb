```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("/content/registro_de_ventas_de_bicicletas_.csv")
print(df.head(10))
print(df.tail(5))
#dimensiones del df
print(df.shape)
```

```python
# Identificar valores nulos
print("\nValores nulos por columna:")
print(df.isnull().sum())
```

```python
# Estadísticas descriptivas
print("\nEstadísticas descriptivas:")
print(df.describe())
```

```python
# Graficar columnas numéricas
col1 = 'Precio '
col2 = 'Cantidad Vendida'
```

```python
# Gráfico de dispersión entre Precio y Cantidad Vendida
plt.scatter(df[col1], df[col2])
plt.xlabel(col1)
plt.ylabel(col2)
plt.title(f'Dispersión entre {col1} y {col2}')
plt.show()
```

```python
# Histograma de Ingresos totales
plt.hist(df['Ingresos totales'], color="pink", bins=10, edgecolor='black') # Added bins and edgecolor
plt.title('Histograma de Ingresos totales')
plt.xlabel('Ingresos totales')
plt.ylabel('Frecuencia')
plt.show()
```

```python

df['Tipo de bici'] = df['Tipo de bici'].astype(str)

plt.bar(df['Tipo de bici'], df['Ingresos totales'])
plt.xlabel('Tipo de bici')
plt.ylabel('Ingresos totales')
plt.title('Número de bicis vendidas')
plt.xticks(rotation=90)
plt.show()
```
