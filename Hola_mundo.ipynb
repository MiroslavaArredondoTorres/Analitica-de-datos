```python
#Importamos las librerias: numpy y matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Definimos 2 arreglos de dimensión 100
ys= 200 + np. random.rand (100)
x= np.arange (0,100,1)
```

```python
#Gráficamos los arreglos en un plano 2D
plt.plot(x,ys, '-')
plt.fill_between(x,ys, 195, facecolor='r', alpha=0.6)

#Agregamos titulo a la gráfica
plt.title("Sample visulization")
plt.xlabel("eje de las x", size=16)
plt.ylabel( "eje de las y", size=16)
plt.show()
```
